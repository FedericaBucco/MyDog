```
def progressBar(value, endvalue, bar_length=30):
    """
    This method is used to display a progress bar for various processes used below
    :param value: current progress value
    :param endvalue: goal value to be reached
    :param bar_length: length of progress bar
    :return
    """

    percent = float(value) / endvalue
    arrow = '-' * int(round(percent * bar_length)-1) + '>'
    spaces = ' ' * (bar_length - len(arrow))
    sys.stdout.write("\rProgress: [{0}] {1}%".format(arrow + spaces, int(round(percent * 100))))
    sys.stdout.flush()
```
